```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("/content/compvis1.webp")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)  #in opencv the colour pattern is in bgr
plt.imshow(img)
plt.title("Step 1:RGB Image")
plt.axis("off")
plt.show()
```

```python
sampeled=cv2.resize(img,(50,50))
plt.imshow(sampeled)
plt.show()
```

```python
print(img.shape)
```

```python
print(img[0:5,0:2])
```

```python
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(gray,cmap="gray")
plt.title("Step2: GrayScale Image")
plt.axis("off")
plt.show()
```

```python
print(gray.shape)
```

```python
print(gray[0:5,0:5])
```

```python
sampled=cv2.resize(gray,(50,50))  #during down scaling few pixels/features will be lost but in up sampling too new wont be added
plt.imshow(sampled,cmap='gray')
plt.title("step 3:sampling(Lower Resolution)")
plt.axis("off")
plt.show()
```

```python
plt.figure(figsize=(4,4))
plt.imshow(sampled,cmap='gray')
plt.grid(True)
plt.title("Pixel Grid")
plt.show()
```

```python
quantized=(gray//64)*64
plt.imshow(quantized,cmap='gray')
plt.axis("off")
plt.show()
```

```python
img=cv2.imread("/content/mustang.jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
print(img.shape)
```

```python
gray=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
plt.imshow(gray,cmap="gray")
plt.show()
```

```python
sampled=cv2.resize(gray,(50,40))
plt.imshow(sampled,cmap='gray')
plt.show()
```

```python
quantize=((gray//8)*8)
plt.imshow(quantize,cmap='Blues')
plt.show()
```
