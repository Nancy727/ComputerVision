```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("/content/sky1.webp")
# Check if the image was loaded successfully
if img is None:
    print("Error: Could not load image. Please check the file path.")
else:
    # Convert the image to grayscale as equalizeHist works on grayscale images
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
equalized_img = cv2.equalizeHist(gray_img)
```

```python
    # Display the original and equalized images
    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)) # Display original in RGB
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(equalized_img, cmap='gray') # Display equalized grayscale image
    plt.title('Histogram Equalized Image')
    plt.axis('off')

    plt.show()
```
