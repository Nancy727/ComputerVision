```python
from tensorflow.keras.datasets import mnist
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
(x_train, _),(_, _)=mnist.load_data()
```

```python
img=x_train[0]
plt.imshow(img,cmap='gray')
plt.title("Original image")
```

Gaussian Noise


```python
noise=np.random.normal(0,30,img.shape)
noisy_gaussian=np.clip(img+noise,0,255).astype(np.uint8)
```

Salt And Pepper Noise

```python
sp_img=img.copy()
prob=0.1
```

```python
sp_img=img.copy()
prob=0.1 #it defines how percentage of the data will be attacked by the error(here 10%)
rand=np.random.rand(*img.shape)
sp_img[rand<prob]=0
sp_img[rand>1-prob]=255
```

```python
mean_img=cv2.blur(noisy_gaussian,(3,3))
```

```python
median_img=cv2.medianBlur(sp_img,3)
```

```python
plt.figure(figsize=(8,8))
plt.subplot(2,2,1)
plt.imshow(noisy_gaussian,cmap='gray')
plt.title("Gaussian Noise")
```

```python
plt.subplot(2,2,2)
plt.imshow(mean_img,cmap='gray')
plt.title("Mean filter")
```

```python
plt.subplot(2,2,3)
plt.imshow(sp_img,cmap='gray')
plt.title("Salt And Pepper Noise")
```

```python
plt.subplot(2,2,2)
plt.imshow(median_img,cmap='gray')
plt.title("Median filter")
```
